```python
# Bibliotecas

# Manipulação de dados
import pandas as pd
import numpy as np

# Gráfica
import matplotlib.pyplot as plt
import seaborn as sns

# SkLearn
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Configurações de plot
plt.style.use('ggplot')
```

```python
dataset = load_breast_cancer()
```

```python
X = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
y = dataset['target']
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape, X_scaled.shape
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(X_pca[:,0], X_pca[:,1], hue=y)
plt.xlabel('Primeiro componente principal')
plt.ylabel('Segundo componente principal')
```

```python
# Componentes da relação linear entre as váriaveis
pca.components_
```

```python
# Num dataframe
df_comp = pd.DataFrame(pca.components_, columns=dataset['feature_names'])
df_comp
```

```python
# Correlações entre parametros e componentes
sns.heatmap(df_comp, cmap='plasma')
```

```python
# Numero de features originais
X.columns.nunique()
```

```python
# Podemos observar quantos % da variancia explicada de acordo com qntde
# de componentes:
# (Vamos testar até o numero de features = 10

pca = PCA(10)
pca.fit(X_scaled)
exp_var = np.cumsum(pca.explained_variance_ratio_)
```

```python
len(exp_var)
```

```python
# Observando relação

sns.lineplot(range(1, 11), exp_var, marker='o')
plt.axhline(0.95, color='blue')
plt.xlabel('Numero de componentes')
plt.ylabel('Variância explicada')
```

```python
# Com cerca de 10 elementos teriamos 95% de variancia explicada
# dos cerca de 30 elementos originais
```
