```python
# Bibliotecas

# Manipulação de dados
import pandas as pd
import numpy as np

# Gráfica
import matplotlib.pyplot as plt
import seaborn as sns

# SkLearn
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

# Configurações de plot
plt.style.use('ggplot')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
X = iris.iloc[:,:-1]
y = iris['species']
```

```python
# Usaremos as classes e compararemos o resultado do KMeans com o resultado das classes originais
```

```python
# Usaremos método do joelho para verificar K ideiais
inertia = []
n = 10
for k in range(1, n):
    kmeans = KMeans(k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
sns.lineplot(range(1,n), inertia, marker='o')
```

```python
# Como podemos perceber o número de K ideais é 3!
```

```python
# Fitando o algoritmo no nosso dataset
kmeans = KMeans(3)
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

```python
# Vamos comparar com algum dos gráficos abaixo
sns.pairplot(iris, hue='species')
```

```python
# Comparando o plot de petal_width x petal_length
sns.scatterplot('petal_width','petal_length', data=iris, hue='species')
# Comparando o plot de petal_width x petal_length
plt.figure()
sns.scatterplot('petal_width','petal_length', data=iris, hue=y_pred,palette='Blues')
```

```python
# Comparando o plot de sepal_width x sepal_length
sns.scatterplot('sepal_width','sepal_length', data=iris, hue='species')
# Comparando o plot de petal_width x petal_length
plt.figure()
sns.scatterplot('sepal_width','sepal_length', data=iris, hue=y_pred,palette='Blues')
```
